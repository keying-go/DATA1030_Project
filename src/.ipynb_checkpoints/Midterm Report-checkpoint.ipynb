```python
import pandas as pd
# load the dataset
df = pd.read_csv("online_shoppers_intention.csv")
```

```python
print(df.shape)
df.head()
```

```python
print(df.dtypes)
```

```python
for col in df.columns:
    print(col)
    print(df[col].value_counts)
    print(df[col].describe())
    print()
```

```python
# Visualize target variable 
import matplotlib
from matplotlib import pylab as plt
pd.value_counts(df["Revenue"],normalize=True).plot.bar()
plt.ylabel('fraction')
plt.xlabel("Made the purchase")
plt.savefig("")
plt.show()
print("purchased count:",sum(df["Revenue"]==True))
print("not purchased count:", sum(df["Revenue"]==False))
```

```python
help(pd.plot.bar)
```

```python
# visualize continuous variable 
df["BounceRates"].plot.hist(bins = 20)
plt.xlabel("BounceRates")
```

```python
df["ExitRates"].plot.hist(bins = 20)
plt.xlabel("ExitRates")
```

```python
df["SpecialDay"].plot.hist(bins = 20)
plt.xlabel("SpecialDay")
```

```python
for col in df.columns[0:-2]:
    # plot categorical columns vs. target
    if df[col].dtypes == "int64" and df[col].nunique() < 20:
        count_matrix = df.groupby([col, "Revenue"]).size().unstack()
        count_matrix_norm = count_matrix.div(count_matrix.sum(axis=1),axis=0)
        print(count_matrix_norm)
        count_matrix_norm.plot(kind='bar', stacked=True)
        plt.ylabel('fraction of people in group')
        plt.legend(loc=4)
    # plot continuous columns vs. target
    elif df[col].dtypes == "float64" or df[col].nunique() >= 20:
        df[[col,"Revenue"]].boxplot(by="Revenue")
        plt.ylabel(col)
plt.show()
```
